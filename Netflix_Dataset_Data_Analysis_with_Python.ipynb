**Que 2:**

*In which year highest number of TV shows & Movies were Released? Plot the Bar Graph*

```python
#Release_Date has Object as its Dtype
data.dtypes
```

```python
#Converting object to datatime format
#By creating a new column
data["Date_N"] = pd.to_datetime(data["Release_Date"])
```

```python
data.head()
```

```python
#Date_N has now data type as datetime64[ns]
data.dtypes
```

```python
#Occurance of Movies and TV shows in each year
data['Date_N'].dt.year.value_counts()
```

```python
#plotting of Bar Graph
data['Date_N'].dt.year.value_counts().plot(kind = "bar")
plt.show()
```
